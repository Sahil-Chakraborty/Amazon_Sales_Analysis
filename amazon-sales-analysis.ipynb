```python
# Basic Setup for Data Analysis in VS Code

#  Core libraries
import pandas as pd
import numpy as np

# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

print(" Environment Ready - Pandas, NumPy, Matplotlib, Seaborn Loaded")
```

```python
my_data = pd.read_csv('/Users/sahilchakraborty/Desktop/Data_analysis/amazon.csv', index_col=0)
print("DAta Loaded succsessfully ")
```

```python
my_data.head(5)
```

**Product Category offering the highest average Discount Percentage**

```python
# Remove '%' and convert to float
my_data['discount_percentage_num'] = my_data['discount_percentage'].str.strip('%').astype(float)

df1 = my_data.groupby('category')['discount_percentage_num'].mean().sort_values(ascending=False)
df1.head()

data=df1.head(10)

plt.title('Product Category offering the highest average Discount Percentage')
sns.barplot(x=data.values,y=data.index,palette='viridis')
plt.xlabel('Product Category')
plt.ylabel('Average discount percentage ')
```

Analysing Wheather Higher discounts links to lower/higher average ratings?

```python
discount_rating = my_data.groupby('discount_percentage_num')['rating'].mean()


plt.figure(figsize=(10,6))
sns.scatterplot(x=discount_rating.index, y=discount_rating.values, color='black')
sns.lineplot(x=discount_rating.index, y=discount_rating.values, color='green', linewidth=2)
plt.title('Discount Percentage vs Average Rating')
plt.xlabel('Discount Percentage')
plt.ylabel('Average Rating')
plt.show()
```

**Analysing Product Category having the highest average rating**

```python
df2=my_data.groupby('category')['rating'].mean()
data3=df2.max()
print(df2.loc[df2 == data3 ])
```

**Comparing average actual price vs discounted price across top 5 categories.**

```python
my_data['Actual_price']= my_data['actual_price'].str.replace('₹','').str.replace(',','').astype(float)
my_data['discounted_price_num']=my_data['discounted_price'].str.replace('₹','').str.replace(',','').astype(float)
df3=my_data.groupby('category')[['Actual_price','discounted_price_num']].mean().sort_values(by ='Actual_price',ascending=False)
df3.head(5)

sns.barplot(x=df3['Actual_price'].head(5), y=df3.index[:5], color='blue')
sns.barplot(x=df3['discounted_price_num'].head(5), y=df3.index[:5], color='orange')
plt.title('Average Actual Price vs Discounted Price across Top 5 Categories')
plt.xlabel('Average Actual Price')
plt.ylabel('Average Discounted Price')
plt.show()
```

**Best selling Products per Category**

```python
best=my_data.loc[my_data.groupby('category')['rating_count'].idxmax()]
best_product = best[['category', 'product_name', 'rating_count']].sort_values(by='rating_count', ascending=False)
print("Best Product based on Rating Count:")
data5=best_product.head(10)
sns.barplot(x=data5['rating_count'],y=data5['category'],hue=data5['product_name'], palette='coolwarm')
plt.legend(title='Product Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Top Products by Rating Count in Each Category')
plt.xlabel('Rating Count')
plt.ylabel('Category')
plt.show()
```

**Conclusion**

From the analysis of Amazon product data:

1. Discount Impact on Ratings: Discount percentage has minimal influence on ratings. Products with both low and high discounts maintain average ratings between 3.9–4.3, suggesting customers value product quality more than discounts.

2. Pricing Trends: Categories such as Split-System Air Conditioners and Traditional Laptops show the largest price drops, highlighting strong deals in premium product ranges.

3. High Discount Categories: Accessories like Cable Protectors, Ear Pads, and Adapters have consistently high discounts (80–90%), often as a competitive sales strategy to drive accessory purchases.

4. Top Products by Popularity: High rating counts are concentrated in essential and affordable items like HDMI Cables, Earphones, and Chargers, indicating these categories drive the bulk of sales volume.

Key Insight for Businesses & Consumers:

* Businesses can focus discounts on low-cost accessories to maintain traffic while keeping premium products attractive through competitive pricing.

* Consumers should monitor categories with consistent high discounts for better value without compromising on quality.
